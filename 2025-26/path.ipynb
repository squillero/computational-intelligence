```python
from random import randint, shuffle
from icecream import ic
```

```python
NUMBERS = [9, 10, 6, 8, 5]
TARGET = 15

NUMBERS = [randint(1, 100) for _ in range(20)]
TARGET = sum(NUMBERS[:5])
shuffle(NUMBERS)
```

## Helpers

```python
def goal_test(state: list[bool]) -> bool:
    return sum(e for e, m in zip(NUMBERS, state) if m) == TARGET


def available_actions(state: list[bool]) -> list[int]:
    return [i for i, e in enumerate(state) if not e]


def do_action(state: list[bool], action: int) -> list[bool]:
    return state[:action] + [True] + state[action + 1 :]


def evaluate_cost(path: list[int]) -> int:
    return len(path)
```

```python
def search(state: list[bool]):
    path: list[int] = list()
    frontier: list[list[bool]] = list()
    cost: dict[tuple[bool, ...], int] = dict()

    while not goal_test(state):
        # ic(state)
        for a in available_actions(state):
            new_state = do_action(state, a)
            c = evaluate_cost(path + [a])
            frontier.append(new_state)
            cost[tuple(new_state)] = c
        # state = frontier.pop(0) # BREADTH FIRST
        state = frontier.pop()  # DEPTH FIRST

    # ic(state)
    print(f"Solution: {[n for n, m in zip(NUMBERS, state) if m]}; sum: {TARGET}")


search([False] * len(NUMBERS))
```
